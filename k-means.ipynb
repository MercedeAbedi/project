```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs 
%matplotlib inline
```

```python
# Create our own dataset
np.random.seed(0)
```

```python
# Making random clusters of points by using the make_blobs class
x, y = make_blobs(n_samples= 5000, centers=[[4,4], [-2, -1], [2, -3], [1, 1]], cluster_std= 0.9)
```

```python
type(x)
```

```python
len(x)
```

```python
len(y)
```

```python
# Display the scatter plot of the randomly generated data
plt.scatter(x[:,0], x[:,1], marker= '.')
```

```python
import seaborn as sns
sns.scatterplot(x= x[:,0], y= x[:,1], hue= y, palette= 'tab10')
```

### K-Means

```python
# Initialize KMeans with these parameters
k_means = KMeans(n_clusters= 4, init= 'k-means++', n_init= 12)
```

```python
# Fit the KMeans model with the feature matrix we created above, x
k_means.fit(x)
```

```python
# Get the coordinates of the cluster centers
center = k_means.cluster_centers_
center
```

```python
# Grab the labels for each point in the model
labels = k_means.labels_
labels
```

```python
# Plot
plt.scatter(x= x[:,0], y= x[:,1], c= labels, cmap= 'rainbow')
plt.scatter(x= center[:, 0], y= center[:, 1], marker= 'x', s= 150)
plt.show()
```
